```python
import os
import zipfile

def zip_folder(folder_path, output_zip):
    with zipfile.ZipFile(output_zip, 'w', zipfile.ZIP_DEFLATED) as zipf:
        for root, dirs, files in os.walk(folder_path):
            for file in files:
                file_path = os.path.join(root, file)
                arcname = os.path.relpath(file_path, folder_path)
                zipf.write(file_path, arcname)

# Esempio di utilizzo:
folder_path = '/kaggle/working/runs/detect/train2'  # Sostituisci con il percorso della tua cartella
output_zip_path = '/kaggle/working/exp.zip'  # Sostituisci con il percorso desiderato per l'archivio

zip_folder(folder_path, output_zip_path)

print(f'La cartella {folder_path} è stata compressa in {output_zip_path}.')
```
